```python
game_names = games_df.set_index('id')['name']
plays_df['gameName'] = plays_df['gameRefId'].apply(lambda x: game_names[x])
```

```python
plays_df.groupby('gameName')['uuid'].count().sort_values(ascending=False)
```

```python
play_players = plays_df['playerScores'].apply(lambda l: [(i['playerRefId'], i['winner']) for i in l])
play_players
```

```python
from collections import defaultdict
opponent_stats = defaultdict(lambda: {'total': 0, 'i_won': 0, 'opponent_won': 0, 'both_lost': 0})

for play in play_players:
    i_won = False
    for player_id, won in play:
        if player_id == 1: #me
            i_won = won
        else:
            opponent_stats[player_id]['total'] += 1
            if i_won:
                opponent_stats[player_id]['i_won'] += 1
            elif won:
                opponent_stats[player_id]['opponent_won'] += 1
            else: 
                opponent_stats[player_id]['both_lost'] += 1

opponent_stats
```

```python
opponent_stats_df = pd.DataFrame(opponent_stats).T.reset_index().rename(columns={'index': 'playerRefId'})
opponent_stats_df['player_name'] = opponent_stats_df['playerRefId'].apply(lambda x: player_names[x])
opponent_stats_df['win_rate'] = round(opponent_stats_df['i_won'] * 100.0 / (opponent_stats_df['i_won'] + opponent_stats_df['opponent_won']), 2)
opponent_stats_df.sort_values(by='win_rate')
```
